```python
import os

dun = set(os.listdir('/s/project/uk_biobank/processed/clumping/'))

df = pd.read_parquet('/s/project/geno2pheno/predictions/bayesian/v108cov_deepRVAT_genes_extended.pq')
todo = set(df.phenocode.unique())

len(dun.intersection(todo))
```

```python
df[['trait', 'phenocode']].drop_duplicates()
```

```python
# Get index variants after clumping

pval_threshold = 5e-8

ind_var_list = []
for pcode in tqdm(set(df.phenocode)):
    ind_vars = pd.read_csv(f'/s/project/uk_biobank/processed/clumping/{pcode}/GWAS_variants.clumped.vars')
    ind_vars.rename(columns={'SNP':'varid'}, inplace=True)
    
    all_vars = pd.read_csv(f'/s/project/uk_biobank/processed/clumping/{pcode}/p_values.tsv', sep='\t')
    
    ind_vars = pd.merge(all_vars, ind_vars, on='varid')
    ind_vars['trait'] = df[df.phenocode==pcode].trait.iloc[0]
    ind_var_list.append(ind_vars[ind_vars.pval_meta_raw <= pval_threshold])
    
ind_dt = pd.concat(ind_var_list)[['chr', 'pos', 'ref', 'alt', 'pval_meta_raw', 'varid', 'trait']].reset_index(drop=True)
ind_dt
```

```python
ind_dt['chr'] = ind_dt['chr'].astype(str)
# ind_dt.to_parquet('/s/project/geno2pheno/data/enrichment_data/index_variants.pq')
```

```python
ind_dt = pd.read_parquet('/s/project/geno2pheno/data/enrichment_data/index_variants.pq')
ind_dt.trait.value_counts()
```

# Find all genes in a radius r of the index variant

```python
# Find all genes within a radius of the index variant

radius = 5e5

temp_dt = ind_dt.copy()

temp_dt['nearby_genes'] = ''

for i, row in tqdm(temp_dt.iterrows()):
    c = str(row['chr'])
    p = row['pos']
    
    nearby_genes = gene_annotation_df[(gene_annotation_df.Chromosome==c) & (gene_annotation_df.Start >= p-radius) & (gene_annotation_df.End <= p+radius)].gene_id
    
    # Store the matching genes in the corresponding row of ind_dt
    temp_dt.at[i, 'nearby_genes'] = ','.join(nearby_genes)

    
temp_dt['radius'] = radius
temp_dt['chr'] = temp_dt['chr'].astype(str)
temp_dt
```

```python
# Write to file
temp_dt.to_parquet(f'/s/project/geno2pheno/data/enrichment_data/genes_near_variants_r_1e{str(int(np.log10(radius)))}.pq')
```

# Find k nearest genes to index variant

```python
k=3

temp_dt = ind_dt.copy()

temp_dt['nearby_genes'] = ''

for i, row in tqdm(temp_dt.iterrows()):
    c = str(row['chr'])
    p = row['pos']

    # Calculate distances from the given position
    gene_annotation_df['Distance'] = abs(gene_annotation_df['Start'] - p)
    
    # Filter genes within the radius and select top k nearest genes
    nearest_genes = gene_annotation_df[(gene_annotation_df['Chromosome'] == c)].nsmallest(k, 'Distance')['gene_id']
                                        # & (gene_annotations_dt['Distance'] <= radius)].nlargest(k, 'Distance')['Gene']
    
    # Store the nearest genes in the corresponding row of ind_dt
    temp_dt.at[i, 'nearby_genes'] = ','.join(nearest_genes)
    
temp_dt['radius'] = f'{k}_NN'
temp_dt['chr'] = temp_dt['chr'].astype(str)
temp_dt
```

```python
# Write to parquet file

# temp_dt['chr'] = temp_dt['chr'].astype(str)
temp_dt.to_parquet(f'/s/project/geno2pheno/data/enrichment_data/genes_near_variants_KNN_{k}.pq')
```
